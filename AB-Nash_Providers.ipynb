```python
#Select Nashville CBSA from zip_cbsa
query = "SELECT * FROM zip_cbsa WHERE cbsa = 34980"

with sqlite3.connect('data/hop_teaming.sqlite') as db: 
    zip_cbsa_sqlite = pd.read_sql(query, db);

print(zip_cbsa_sqlite.shape)
zip_cbsa_sqlite.head()
```

```python
start = time.time()

query = '''
SELECT * 
FROM npi 
INNER JOIN zip_cbsa
USING(zip)
INNER JOIN taxonomy
ON taxonomy.code = npi.healthcare_provider_taxonomy_code
WHERE cbsa = 34980
'''
with sqlite3.connect('data/hop_teaming.sqlite') as db: 
    npi_nash_sqlite = pd.read_sql(query, db);

end = time.time()    

print(f'The query took {end-start} seconds to run')
print(npi_nash_sqlite.shape)
npi_nash_sqlite.head()
```

```python
start = time.time()

query = '''
SELECT * 
FROM npi 
INNER JOIN zip_cbsa
USING(zip)
INNER JOIN taxonomy
ON taxonomy.code = npi.healthcare_provider_taxonomy_code
INNER JOIN hopteaming
ON hopteaming.to_npi = npi.npi
WHERE cbsa = 34980
'''
with sqlite3.connect('data/hop_teaming.sqlite') as db: 
    to_npi_nash_sqlite = pd.read_sql(query, db);

end = time.time()    

print(f'The query took {end-start} seconds to run')
print(to_npi_nash_sqlite.shape)
to_npi_nash_sqlite.head()
```
