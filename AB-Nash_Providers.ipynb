```python
start = time.time()

query = '''
WITH dig5zip_npi AS(
SELECT *, CAST(SUBSTR(provider_business_practice_location_address_postal_code,1,5) AS INT) AS zip 
FROM npi) 

SELECT * 
FROM dig5zip_npi 
INNER JOIN zip_cbsa
USING(zip)
INNER JOIN taxonomy
ON taxonomy.code = dig5zip_npi.healthcare_provider_taxonomy_code
WHERE cbsa = 34980
'''
with sqlite3.connect('data/hop_teaming.sqlite') as db: 
    npi_nash_sqlite = pd.read_sql(query, db);

end = time.time()    

print(f'The query took {end-start} seconds to run')
print(npi_nash_sqlite.shape)
npi_nash_sqlite.head()
```

```python
start = time.time()

query = '''
WITH dig5zip_npi AS(
SELECT *, CAST(SUBSTR(provider_business_practice_location_address_postal_code,1,5) AS INT) AS zip 
FROM npi) 

SELECT * 
FROM dig5zip_npi 
INNER JOIN zip_cbsa
USING(zip)
INNER JOIN taxonomy
ON taxonomy.code = dig5zip_npi.healthcare_provider_taxonomy_code
INNER JOIN hopteaming
ON hopteaming.to_npi = dig5zip_npi.npi
WHERE cbsa = 34980
'''
with sqlite3.connect('data/hop_teaming.sqlite') as db: 
    to_npi_nash_sqlite = pd.read_sql(query, db);

end = time.time()    

print(f'The query took {end-start} seconds to run')
print(to_npi_nash_sqlite.shape)
to_npi_nash_sqlite.head()
```
