```python
start = time.time()

query = '''
SELECT * 
FROM npi 
INNER JOIN zip_cbsa
USING(zip)
INNER JOIN taxonomy
ON taxonomy.code = npi.healthcare_provider_taxonomy_code
WHERE cbsa = 34980
'''
with sqlite3.connect('data/hop_teaming.sqlite') as db: 
    npi_nash_sqlite = pd.read_sql(query, db);

end = time.time()    

print(f'The query took {end-start} seconds to run')
print(npi_nash_sqlite.shape)
npi_nash_sqlite.head()
```

```python
start = time.time()

query = '''
SELECT * 
FROM npi 
INNER JOIN zip_cbsa
USING(zip)
INNER JOIN taxonomy
ON taxonomy.code = npi.healthcare_provider_taxonomy_code
INNER JOIN hopteaming
ON hopteaming.to_npi = npi.npi
WHERE cbsa = 34980
'''
with sqlite3.connect('data/hop_teaming.sqlite') as db: 
    to_npi_nash_sqlite = pd.read_sql(query, db);

end = time.time()    

print(f'The query took {end-start} seconds to run')
print(to_npi_nash_sqlite.shape)
to_npi_nash_sqlite.head()
```

```python
start = time.time()

query = '''
SELECT * 
FROM npi 
INNER JOIN zip_cbsa
USING(zip)
INNER JOIN taxonomy
ON taxonomy.code = npi.healthcare_provider_taxonomy_code
INNER JOIN hopteaming
ON hopteaming.from_npi = npi.npi
WHERE cbsa = 34980
'''
with sqlite3.connect('data/hop_teaming.sqlite') as db: 
    from_npi_nash_sqlite = pd.read_sql(query, db);

end = time.time()    

print(f'The query took {end-start} seconds to run')
print(from_npi_nash_sqlite.shape)
from_npi_nash_sqlite.head()
```

```python
to_npi_nash_sqlite.groupby(['provider_organization_name_(legal_business_name)'])['provider_organization_name_(legal_business_name)'].count().nlargest(50)
```

```python
nash_hospitals = to_npi_nash_sqlite[to_npi_nash_sqlite['grouping'] == 'Hospital Units']
```

```python
nash_hospitals.groupby(['provider_organization_name_(legal_business_name)'])['provider_organization_name_(legal_business_name)'].count().nlargest(50)
```

```python
nash_hospitals_npi_dict = nash_hospitals[['npi','provider_organization_name_(legal_business_name)']].drop_duplicates().set_index('npi').to_dict()
nash_hospitals_npi_dict
```

```python
nash_hospitals_npi_list = nash_hospitals['npi'].unique().tolist()
nash_hospitals_npi_list
```

```python
nash_hospital_referrals = from_npi_nash_sqlite[from_npi_nash_sqlite['to_npi'].isin(nash_hospitals_npi_list)].copy()
nash_hospital_referrals
```

```python
nash_hospital_referrals = nash_hospital_referrals.merge(nash_hospitals[['to_npi','provider_organization_name_(legal_business_name)']].rename(columns={'provider_organization_name_(legal_business_name)': 'to_hospital_name'}),how='inner', on='to_npi').drop_duplicates()
```

```python
nash_hospital_referrals['from_name-to_name'] = nash_hospital_referrals['provider_organization_name_(legal_business_name)'] + ' - ' + nash_hospital_referrals['to_hospital_name']
```

```python
nash_hospital_referrals.groupby(['provider_organization_name_(legal_business_name)'])['provider_organization_name_(legal_business_name)'].count().nlargest(50)
```

```python
nash_hospital_referrals.groupby(['provider_organization_name_(legal_business_name)','specialization'])['patient_count'].sum().nlargest(50)
```

```python
to_name_from_name_referral_sums = nash_hospital_referrals.groupby(['from_name-to_name','specialization'])['patient_count'].sum().reset_index().sort_values('patient_count', ascending = False)
to_name_from_name_referral_sums
```

```python
specialization_list = nash_hospital_referrals['specialization'].unique().tolist()
specialization_list
```

```python
to_hospital_referral_sums = nash_hospital_referrals.groupby(['to_hospital_name','specialization'])['patient_count'].sum().reset_index().sort_values('patient_count', ascending = False)
to_hospital_referral_sums
```

```python
#dict comprehension for the to_hospital patient_count sums by specialty
dict_to_hospital_name_by_specialty = {specialty: to_hospital_referral_sums[to_hospital_referral_sums['specialization'] == specialty] for specialty in specialization_list }
```

```python
dict_to_hospital_name_by_specialty['Adult Health'].sort_values('patient_count', ascending=False)
```

```python
#dict comprehension for the to_hospital patient_count sums by specialty
dict_to_name_from_name_by_specialty = {specialty: to_name_from_name_referral_sums[to_name_from_name_referral_sums['specialization'] == specialty] for specialty in specialization_list}
```

```python
dict_to_name_from_name_by_specialty['Diagnostic Radiology'].sort_values('patient_count', ascending=False)
```

```python
to_hospital_referral_sums['pct_patient_count_by_specialization'] = to_hospital_referral_sums['patient_count'] / to_hospital_referral_sums.groupby('specialization')['patient_count'].transform('sum')
```

```python
fig = px.bar(to_hospital_referral_sums, y="specialization", x="pct_patient_count_by_specialization", color="to_hospital_name", title="Hospital Patient Referrals By Specialty", height = 1000)
fig.show()
```
