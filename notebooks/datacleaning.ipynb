```python
import pandas as pd
import sqlite3
```

```python
database_path = '../data/hop_teaming'
```

```python
query='''
SELECT * 
FROM nash_referrals
'''
with sqlite3.connect(database_path) as db:
    referrals = pd.read_sql(query, db)
```

```python
print(referrals.shape)
referrals.head()
```

```python
query ="""
SELECT 
    DISTINCT nr.referrer_npi as referrer_npi, 
    npi.entity_type_code as referrer_entity_type_code,
    npi.provider_last_name as referrer_last_name, 
    npi.provider_first_name as referrer_first_name, 
    npi.provider_credential_text as referrer_credential,
    tx.Specialization as referrer_specialization, 
    npi.provider_zip_code as referrer_zip_code,
    npi.provider_business_practice_location_address_city_name as referrer_city,
    npi.provider_business_practice_location_address_state_name as referrer_state
FROM 
    nash_referrals AS nr
    INNER JOIN npi ON  nr.referrer_npi  = npi.npi
    INNER JOIN taxonomy AS tx ON npi.taxo_code = tx.Code
WHERE
    npi.entity_type_code = '1'
"""
with sqlite3.connect(database_path) as db:
    referring_providers1 = pd.read_sql(query, db)
```

```python
query ="""
SELECT 
    DISTINCT nr.referrer_npi as referrer_npi, 
    npi.entity_type_code as referrer_entity_type_code,
    npi.provider_business_name AS referrer_business_name, 
    tx.Grouping AS referrer_grouping, 
    tx.Classification as referrer_classification,
    npi.provider_zip_code as referrer_zip_code,
    npi.provider_business_practice_location_address_city_name as referrer_city,
    npi.provider_business_practice_location_address_state_name as referrer_state
FROM 
    nash_referrals AS nr
    INNER JOIN npi ON  nr.referrer_npi  = npi.npi
    INNER JOIN taxonomy AS tx ON npi.taxo_code = tx.Code
WHERE
    npi.entity_type_code = '2'
"""

with sqlite3.connect(database_path) as db:
    referring_providers2 = pd.read_sql(query, db)
```

```python
referrals.to_csv('../clean_data/referrals.csv', index=False)
referring_providers2.to_csv('../clean_data/referring_providers2.csv', index=False)
referring_providers1.to_csv('../clean_data/referring_providers1.csv', index=False)
```
