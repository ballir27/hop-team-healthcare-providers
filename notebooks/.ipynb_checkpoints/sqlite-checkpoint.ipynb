**Goal: Create a SQL database for the four datasets.**

**#1 hopteaming dataset**

**Remove "accidental" referrals in hopteaming DataFrame.** 

Filter for records where transaction_count is at least 50 and the average_day_wait is less than 50.

```python
db = sqlite3.connect('../data/hop_teaming.sqlite')

for chunk in pd.read_csv('../data/DocGraph_Hop_Teaming_2018.csv', 
                        chunksize = 100000):
    # Apply the filter 
    filter_condition = (chunk['transaction_count'] > 50) & (chunk['average_day_wait'] < 50)
    filtered_chunk = chunk[filter_condition]
    
    # Fixes the Index error 
    filtered_chunk = filtered_chunk.reset_index(drop = True)
    
    # Clean up the column names
    filtered_chunk.columns = [x.lower().replace(' ', '_') for x in filtered_chunk.columns]    
    
    # Append the chunk to hop_teaming table
    filtered_chunk.to_sql('hopteaming', 
                 db, 
                 if_exists = 'append', 
                 index = False)

    print(f"{len(filtered_chunk)} rows added")
    
db.close()
```

**#2 Taxonomy Dataset**

```python
db = sqlite3.connect('../data/hop_teaming.sqlite')

for chunk in pd.read_csv('../data/nucc_taxonomy_240.csv', 
                        chunksize = 100000):
    chunk.columns = [x.lower().replace(' ', '_') for x in chunk.columns] 
    chunk.to_sql('taxonomy', 
                 db, 
                 if_exists = 'append', 
                 index = False) 
    
    print(f"{len(chunk)} rows added")

print("Done")
result = db.execute('SELECT COUNT(*) FROM taxonomy')
total_rows_taxonomy = result.fetchone()[0]
print(f"Total rows: {total_rows_taxonomy}")
    
db.commit()
db.close()
```

**#4 NPPES (NPI) Dataset**

**Select the relevant columns in NPPES dataset for this project.**

- 'NPI'
- Entity Type, indicated by the 'Entity Type Code' field:
    - 1 = Provider (doctors, nurses, etc.)
    - 2 = Facility (Hospitals, Urgent Care, Doctors Offices)
- Entity Name: Either First/Last or Organization or Other Organization Name contained in the following fields:
    - 'Provider Organization Name (Legal Business Name)'
    - 'Provider Last Name (Legal Name)'
    - 'Provider First Name'
    - 'Provider Middle Name'
    - 'Provider Name Prefix Text'
    - 'Provider Name Suffix Text'
    - 'Provider Credential Text'
- Address: Business Practice Location (not mailing), contained in the following fields:
    - 'Provider First Line Business Practice Location Address'
    - 'Provider Second Line Business Practice Location Address'
    - 'Provider Business Practice Location Address City Name'
    - 'Provider Business Practice Location Address State Name'
    - 'Provider Business Practice Location Address Postal Code'
- The provider's taxonomy code, which is contained in one of the 'Healthcare Provider Taxonomy Code*' columns.

**Retrieve the provider's taxonomy code from the npidata DataFrame.**

A provider can have up to 15 taxonomy codes, but we want the one which has Primary Switch = Y in the associated 'Healthcare Provider Primary Taxonomy Switch*' field. Note that this does not always occur in spot 1.

```python
db = sqlite3.connect('../data/hop_teaming.sqlite')

for chunk in pd.read_csv('../data/npidata_pfile_20050523-20240211.csv', 
                         low_memory = False,
                         chunksize = 100000):
    # Apply the filter 
    filter_columns = ['NPI',
                      'Entity Type Code',
                      'Provider Organization Name (Legal Business Name)',
                      'Provider Last Name (Legal Name)',
                      'Provider First Name',
                      'Provider Middle Name',
                      'Provider Name Prefix Text',
                      'Provider Name Suffix Text',
                      'Provider Credential Text',
                      'Provider First Line Business Practice Location Address',
                      'Provider Second Line Business Practice Location Address',
                      'Provider Business Practice Location Address City Name',
                      'Provider Business Practice Location Address Postal Code']
    filtered_chunk = chunk[filter_columns]
    
    # Clean up the column names
    filtered_chunk.columns = [x.lower().replace(' ', '_') for x in filtered_chunk.columns] 
    
    # Reset the index 
    filtered_chunk = filtered_chunk.reset_index(drop=True)
    
    # Create a new column for Taxonomy Code 
    filtered_chunk['Healthcare Provider Taxonomy Code'] = np.nan

    # Iterrate over rows to fill the Taxonomy Code column 
    for index, row in chunk.iterrows():
        for i in range(1, 16):  # For the 15 columns relating to taxonomy code
            code_col = f'Healthcare Provider Taxonomy Code_{i}'
            switch_col = f'Healthcare Provider Primary Taxonomy Switch_{i}'
            # Check if the taxonomy switch is 'Y' and the code is not NaN
            if row[switch_col] == 'Y' and not pd.isna(row[code_col]):
                filtered_chunk.at[index, 'Healthcare Provider Taxonomy Code'] = row[code_col]
                
    # Append the chunk to a hop_teaming table
    filtered_chunk.to_sql('npi', 
                 db, 
                 if_exists = 'append', 
                 index = False)  
    
    print(f"{len(filtered_chunk)} rows added")

print("Done")
result = db.execute('SELECT COUNT(*) FROM npi')
total_rows_npi = result.fetchone()[0]
print(f"Total rows: {total_rows_npi}")

db.commit()
db.close()
```
